```python
#Criar uma funcao para pegar uma amostra sistemÃ¡tica

def amostragem_sistematica(df, step):
  indexes = np.arange(3, len(df), step=step)
  amostra_sistematica = iris_df.iloc[indexes]
  return(amostra_sistematica)

amostra_sistemataca= amostragem_sistematica(iris_df,3)

display(amostra_sistemataca ) 
  
```
